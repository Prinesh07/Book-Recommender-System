### Popularity based Recommender System

```python
books_with_ratings = ratings_df.merge(book_df, on = 'ISBN')
books_with_ratings.head()
```

```python
num_ratings_df = books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.columns = ['Book-Title', 'num_ratings']
num_ratings_df
```

## Collaborative filtering based recommendation system

```python
top_users = books_with_ratings.groupby('User-ID').count()['Book-Rating'] > 200
top_users = top_users[top_users].index
top_users
```

```python
filtered_ratings = books_with_ratings[books_with_ratings['User-ID'].isin(top_users)]
filtered_ratings
```

```python
famous_books = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = famous_books[famous_books].index
famous_books
```

```python
top_users_and_top_books = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
top_users_and_top_books
```

```python
users_and_books_pivot_table = top_users_and_top_books.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
users_and_books_pivot_table.fillna(0, inplace = True)
users_and_books_pivot_table
```

### Recommendation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(users_and_books_pivot_table)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # Fetch the index of the book from the pivot table
    index = np.where(users_and_books_pivot_table.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])), key = lambda x : x[1], reverse = True)[1:6]

    recommended_books = []
    for i in similar_books:
        items = []
        temp_df = book_df[book_df['Book-Title'] == users_and_books_pivot_table.index[i[0]]]
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        items.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))

        recommended_books.append(items)

    return recommended_books
```

```python
recommend('A Man Named Dave: A Story of Triumph and Forgiveness')
```
